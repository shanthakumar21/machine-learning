```python
#making directories for positive and negative
import os
dir = os.path.join("Assignment_1_a/positive")
if not os.path.exists(dir):
    os.mkdir(dir)
    
dir = os.path.join("Assignment_1_a/negative")
if not os.path.exists(dir):
    os.mkdir(dir) 
```

```python
#copying the positive reviews
with open('Assignment_1_a/Positive_Reviews.txt', 'r') as f:
    mylist = ast.literal_eval(f.read())
    for i in mylist:
        source ='Assignment_1_a/Reviews/'+i
        des = 'Assignment_1_a/positive'
        shutil.copy(source, des)
```

```python
with open('Assignment_1_a/Negative_Reviews.txt', 'r') as f1:
    mylist = ast.literal_eval(f1.read())
    for i in mylist:
        source = 'Assignment_1_a/Reviews/'+i
        des ='Assignment_1_a/negative'
        shutil.copy(source, des)
```

```python
with open(txt_file, 'r') as infile, open(csv_file, 'w') as outfile:
    stripped = (line.strip() for line in infile)
    lines = (line.split(" ") for line in stripped if line)
    writer = csv.writer(outfile)
    writer.writerows(lines)
```
